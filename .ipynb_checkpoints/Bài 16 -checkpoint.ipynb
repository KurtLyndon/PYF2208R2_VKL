```python
class Battery():
    energy = ""
    def __init__(self):
        self.energy = 10
        
    def setEnergy(self, energy = 10):
        self.energy = energy
        
    def getEnergy(self):
        print(f"Energy left = {self.energy}")
        
    def decreaseEnergy(self):
        self.energy = self.energy - 2
        if(self.energy < 0):
            self.energy = 0
    
    
class FlashLamp():
    status = False
    battery = Battery()
    
    def __init__(self):
        self.status = False
        battery = Battery()
        
    def setBattery(self, battery):
        self.battery.setEnergy(battery.energy)
    
    def getBatteryInfo(self):
        self.battery.getEnergy()
        
    def turnOn(self):
        print("Try to turn on")
        self.status = True
        self.battery.decreaseEnergy()
        if(self.battery.energy > 0):
            print("Light on")
        else:
            print("Light off")
            
    def turnOff(self):
        print("Turn off")
        self.status = False

b = Battery()
b.setEnergy(15)
flash = FlashLamp()
flash.setBattery(b)
for i in range(10):
    print(f"-----\nAttemp {i+1}")
    flash.turnOn()
    flash.getBatteryInfo()
    flash.turnOff()
    flash.getBatteryInfo()
```
