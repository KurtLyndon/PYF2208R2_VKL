```python
from tkinter import *
from tkinter import messagebox

class ingredient():
    name = ""
    quantity = 0
    
    def __init__(self, a, b):
        self.name = a
        self.quantity = b
    def add(self, a = "", b = 0):
        self.name = a
        self.quantity = b
        
window = Tk()
window.title("Ingredient Management System")

name_lbl = Label(window, text = "Name:")
quantity_lbl = Label(window, text = "Quantity:")
search_lbl = Label(window, text = "Search:")

name_entry = Entry(window)
quantity_entry = Entry(window)
search_entry = Entry(window)

ingredient_ls = []

def insert_function():
    temp = ingredient(name_entry.get(),int(quantity_entry.get()))
    ingredient_ls.append(temp)
    print("List = {", end="")
    for i in ingredient_ls:
        print(i.name, end=",")
    print("}")
    messagebox.showinfo("Insert Ingredient", f"Insert {name_entry.get()} with quantity: {quantity_entry.get()} completed!")
    name_entry.delete(0,"end")
    quantity_entry.delete(0,"end")
    
def delete_function():
    name_entry.config(state= "normal")
    quantity_entry.config(state= "normal")
    name_entry.delete(0,"end")
    quantity_entry.delete(0,"end")
    flag = -1
    for i in range(0,len(ingredient_ls)):
        if(ingredient_ls[i].name == search_entry.get()):
            flag = i
            break
    
    if(flag != -1):
        messagebox.showinfo("Delete Ingredient", f"{search_entry.get()} deleted from the list!")
        ingredient_ls.pop(flag)
    else:
        messagebox.showinfo("Delete Ingredient", f"{search_entry.get()} not in the list!")

def reset_function():
    name_entry.config(state= "normal")
    quantity_entry.config(state= "normal")
    name_entry.delete(0,"end")
    quantity_entry.delete(0,"end")
    search_entry.delete(0,"end")
    
def search_function():
    name_entry.config(state= "normal")
    quantity_entry.config(state= "normal")
    name_entry.delete(0,"end")
    quantity_entry.delete(0,"end")
    flag = -1
    for i in range(0,len(ingredient_ls)):
        if(ingredient_ls[i].name == search_entry.get()):
            flag = i
            break
    
    if(flag != -1):
        name_entry.insert(0,ingredient_ls[flag].name)
        quantity_entry.insert(0,ingredient_ls[flag].quantity)
        name_entry.config(state= "disabled")
        quantity_entry.config(state= "disabled")
    else:
        messagebox.showinfo("Search Ingredient", f"{search_entry.get()} not in the list!")
    
add_btn = Button(window, text = "Add", command = insert_function)
delete_btn = Button(window, text = "Delete", command = delete_function)
reset_btn = Button(window, text = "Reset", command = reset_function)
search_btn = Button(window, text = "Search", command = search_function)

name_lbl.grid(row = 0, column = 0)
name_entry.grid(row = 0, column = 1, columnspan = 2)

quantity_lbl.grid(row = 1, column = 0)
quantity_entry.grid(row = 1, column = 1)

add_btn.grid(row = 2, column = 0)
reset_btn.grid(row = 2, column = 1)

search_lbl.grid(row = 3, column = 0)
search_entry.grid(row = 3, column = 1)

search_btn.grid(row = 4, column = 0)
delete_btn.grid(row = 4, column = 1)

window.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox
class Menu():
    def __init__(self, foodname, price):
        self.foodname = foodname
        self.price = price
wds = Tk()
wds.title('MENU MANAGEMENT OF YARA RESTAURANT')
wds.geometry("500x300")
list_Foods = []

foodname_lbl = Label(wds, text = "Food Name")
price_lbl = Label(wds, text = "Price")

foodname = Entry(wds)
price = Entry(wds)
check = Entry(wds)

def Add_fooditem():
    list_Foods.append(Menu(foodname.get(), int(price.get())))
    print([item.foodname for item in list_Foods])
    messagebox.showinfo(foodname.get() + " is added successfully")

add_btn = Button(wds, text = "Add", command=Add_fooditem)
delete_btn = Button(wds, text = "Delete")
edit_btn = Button(wds, text = "Edit")
search_btn = Button(wds, text = "Search")

foodname_lbl.place(x=10, y=20)
foodname.place(x=100, y=20)
price_lbl.place(x=10, y=50)
price.place(x=100, y=50)
add_btn.place(x=60, y=90)
delete_btn.place(x=140, y=90)
edit_btn.place(x= 230, y=90)
check.place(x=10, y=140)
search_btn.place(x=160, y=140)

wds.mainloop()

```
